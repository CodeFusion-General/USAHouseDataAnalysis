```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Veri setini pandas ile içe aktar
file_path = r'E:\DijiKent\USAHouseDataAnalysis\Data\HouseDataNew.csv'
house_data = pd.read_csv(file_path)

# Veri seti hakkında genel bilgi edin
print("\nVeri Seti Özeti:")
house_data.info()

# İlk 5 satırı görüntüle
print("\nİlk 5 Satır:")
print(house_data.head())

# Eksik veri analizi
print("\nEksik Veri Sayısı:")
print(house_data.isnull().sum())

# Temel istatistiksel analiz
print("\nTemel İstatistiksel Analiz:")
print(house_data.describe())
```

```python

# Kategorik sütunların dağılımını inceleme
categorical_columns = house_data.select_dtypes(include=['object']).columns
print("\nKategorik Sütunların Değer Dağılımı:")
for col in categorical_columns:
    print(f"\n{col} sütununun değer dağılımı:")
    print(house_data[col].value_counts())

# Sayısal veriler arasındaki korelasyonu inceleme
print("\nSayısal Veriler Arasındaki Korelasyon:")
correlation_matrix = house_data.corr()
print(correlation_matrix)

# Korelasyon matrisini görselleştirme
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Korelasyon Matrisi')
plt.show()

# Fiyat sütununun dağılımını inceleme
plt.figure(figsize=(10, 6))
plt.hist(house_data['price'].dropna(), bins=50, color='blue', alpha=0.7)
plt.title('Fiyat Dağılımı')
plt.xlabel('Fiyat')
plt.ylabel('Frekans')
plt.show()

# Fiyat kategorisine göre ev büyüklüğünün ortalamasını inceleme
avg_size_by_price_category = house_data.groupby('price_category')['house_size'].mean()
print("\nFiyat Kategorisine Göre Ev Büyüklüğünün Ortalaması:")
print(avg_size_by_price_category)

# Fiyat kategorisine göre ev büyüklüğünü görselleştirme
avg_size_by_price_category.plot(kind='bar', figsize=(8, 5), color='orange', alpha=0.7)
plt.title('Fiyat Kategorisine Göre Ortalama Ev Büyüklüğü')
plt.xlabel('Fiyat Kategorisi')
plt.ylabel('Ortalama Ev Büyüklüğü')
plt.show()
```
